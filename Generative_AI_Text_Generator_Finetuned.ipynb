```python
# Step 2: Import necessary modules
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
```

```python
# Step 3: Set device
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
# Step 4: Load a powerful generative AI model
# GPT-J 6B is excellent for high-quality text generation
model_name = "EleutherAI/gpt-j-6B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16 if device=="cuda" else torch.float32)
model.to(device)
```

```python
# Step 5: Define a text generation function
def generate_text(prompt, max_length=300, temperature=0.7, top_k=50, top_p=0.95):
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)
```

```python
 output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
```

```python
text = tokenizer.decode(output[0], skip_special_tokens=True)
    return text
```

```python
# Step 6: Interactive prompt input in Colab
prompt = input("Enter your prompt: ")
generated_text = generate_text(prompt)
print("\nGenerated Text:\n")
print(generated_text)
```
